Understanding Exceptions


Exception Handling is python allows you to handle errors gracefully and take correction actions without stopping the execution of the program . this lesson will cover the basics of exceptions including how to use try except else and finally blocks.


What are the exceptions ?

Exceptions are events that disrupt the normal flow of the program they occur when an error is encountered dureing program execution ,common exception include:


ZeroDivisionError :divinding by zero
FileNoFoundError: file not found
ValueError : Invalid value
TypeError: Invalid type
NameError : varibale not defined 


```python
### Exception try,except block


try:
    a=b
except:
    print("The varible is not defined")
```

```python
# Name Error

try:
    a=b
except NameError as ex:
    print(ex)
```

```python
# Divideby zero exception

try:
    result=1/0
except ZeroDivisionError as ex:
    print("Please enter the denominator greater than 0")
```

```python
try:
    num= int(input("enter the number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")
except Exception as ex:
    print(ex)
```

```python
###  try ,except, else block

try:
    num= int(input("enter the number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")
except Exception as ex:
    print(ex)
else:
    print("Result ",result)
```

```python
### try,except,else and finally

try:
    num= int(input("enter the number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")
except Exception as ex:
    print(ex)
else:
    print("Result ",result)
finally:
    print("Execution complete")
```
