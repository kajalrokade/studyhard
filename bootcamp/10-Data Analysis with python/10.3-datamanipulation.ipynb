```python
### Renaming column  
# give a key value pair 

df=df.rename(columns={'Subscription Date':'Admitted Date'})
df.head(2)
```

```python
## Change the data type
df['Value']=1000
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df['Value_New']=df['Value'].astype(float)

df.dtypes
```

```python
df['Value_New'] = df['Value'].apply(lambda x:x*2)
df.head(5)
```

```python
### Data Aggregating and Grouping

grouped_mean=df.groupby('Country')['Value'].mean()
print(grouped_mean)
```

```python
grouped2=df.groupby(['Country','Admitted Date'])['Value'].sum()
print(grouped2)
```

```python
### Multiple Agregation operation Together

grouped_agg=df.groupby(['Country'])['Value'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
 ### Merging and Joining Dataframes

#create sample dataframes

df1=pd.DataFrame({'Keys':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Keys':['A','B','D'],'Value2':[4,5,6]})

df1
```

```python
df2
```

```python
## Merge DataFrame on the key Column

pd.merge(df1,df2,on='Keys',how='inner')
```

```python
pd.merge(df1,df2,on='Keys',how='outer')
```

```python
pd.merge(df1,df2,on='Keys',how='left')
```

```python
pd.merge(df1,df2,on='Keys',how='right')
```
