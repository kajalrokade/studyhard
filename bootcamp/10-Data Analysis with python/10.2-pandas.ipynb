Pandas- Data Frame and Series


pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. it provided two primary data structures. series and dataframe.
A series is a one dimentional array like object 
while a data frame is a two dimentional ,size mutable and potentionally heterogeneous tabular data structure with labled axes (rows and columns)

```python
import pandas as pd
```

```python
## Series
## Series is one-dimantional array like object that can hold any data type. It is similar to a column in table.



data=[1,2,3,4,5]

series=pd.Series(data)


print("Series: \n",series)
```

```python
## Create a series from a dictionary elements

data={"a":1,"b":2,"c":3}

print(pd.Series(data))
```

```python
data=[10,20,30]
index=['a','b','c']

pd.Series(data,index)
```

```python
## Data Frame

## Create a dataframe from dictionary of list


data={
    "Name":['Kajal','Jon',"jack"],
    "Age":[25,30,45],
    "City":['Pune','Mumbai','Nashik']
}

df=pd.DataFrame(data)
print(df)
print(type(df))


```

```python
## Create a data frame from list of dictionaries

data=[
    {"Name":'Kajal',"Age":25,"City":'Pune'},
    {"Name":'Jon',"Age":25,"City":'Mumbai'},
    {"Name":'JAck',"Age":25,"City":'Nashik'},
    {"Name":'Mac',"Age":25,"City":'Pune'},
    {"Name":'PQR',"Age":25,"City":'Pune'}
]

df=pd.DataFrame(data)
print(df)
print(type(df))

```
