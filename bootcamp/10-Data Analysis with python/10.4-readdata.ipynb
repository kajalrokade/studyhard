Reading Data from Diffrent Sources


```python
import pandas as pd
from io import StringIO
```

```python
Data='{ "name":"Kajal", "email":"kajal@gmail.com", "job_profile": [{"Title":"Team Lead","Salary":20000 }] }'
df=pd.read_json(StringIO(Data))
```

```python
df
```

```python
df.to_json()

## here the default parameter is orient here which id by defaut indec the o is the index of the records
#  similar to the 
# df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```

```python
df.to_json(orient='values')
```
