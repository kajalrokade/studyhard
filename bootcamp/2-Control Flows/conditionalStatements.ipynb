Conditional Statements (if , elif, else)



```python
## if statements

age=20

if age>=18:
    print("You are allowed to vote in election")
```

```python
## else statement

## The else staement executes a block of code if the condition in the if statement is false
 
age=16
if age>=18:
    print("You are eligible for vote")
else:
    print("You are a minor")
```

```python
## elif statement

# The elif statements allows you to check multiple condiitons , it stands for else if

age=17
if age<13:
    print("Ypu are child")
elif age<18:
    print("Y0u are teenager")
else:
    print("You are an adult")
```
