```python
## Loops


# For loops

for i in range(5):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
str="Kajal rokade"
for i in str:
    print(i)
```

```python
## While loops
#  The while loops continues to execute as long as the condition is True

count=0
while count<5:
    print(count)
    count=count+1
```

```python
# Loop Control Statements

# break

for i in range(10):
    if i==5:
        break #
    print(i)
```

```python
# continue

## continue statements skips the current iteration and continues with the next


for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass 
# The pass statement is a null operation , it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loops

# a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
# Example 
# calculate the usm of first n natural numbers


n=10
sum=0

count=1

while(count<=n):
    sum=sum+count
    count=count+1

print("sum of n natural numbers :",sum)
```

```python
# # Example : prime numbers between 1 and 100



for num in range(1,100):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
