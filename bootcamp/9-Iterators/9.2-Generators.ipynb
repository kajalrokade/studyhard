Generators

generators are simpler way to create iterators, they use the yield keyword to produce a series of values lazily which means they generate values on the fly and do not store them in memory.


```python
def square(n):
    for i in range(n):
        yield i**2

    
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a=square(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

    
```

```python
gen =my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

Practical Example : Reading Large Files

Generators are particularly useful for reading large files because they allow you to process one line at a time without loading the entire file into memory.

```python
### Reading large files

def read_large_files(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line

            
```

```python
for line in read_large_files('large_file.txt'):
    print(line.strip())
```
