Iterators

iterators are advanced python concepts that allow for efficient looping and memory  managements.
Iterators provide away to access eleemtns of collection sequntially without exposing the underlying structure.


```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
#  Iterator
iterator=iter(my_list)
print(type(iterator))
```

```python
## It uses a lazy loading
## Iterate through all the elements

next(iterator)
```

```python
iterator=iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("there are no elements in iterator")
```
