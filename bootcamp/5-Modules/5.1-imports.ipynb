## Importing modules in Python : Modules and PAckages



```python
import math
math.sqrt(4)
```

```python
from math import sqrt,pi

print(sqrt(36))
print(pi)
```

```python
# mostly used with arrays and operation on array
import numpy as np

arr= np.array([1,2,3,4,5])
print(arr)
```

```python
# import all the functions from the module


from math import *

print(sqrt(4))
print(pi)
```

```python
# imported a custom package

from package.maths import addition

addition(2,5)
```

```python
# another way call custom package

from package import maths

maths.addition(4,6)
```
