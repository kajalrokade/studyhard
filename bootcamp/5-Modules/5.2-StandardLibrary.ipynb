Standard Library Overview

Pythons standard library is a vast collection of modules and packages that come bundled with python , providing a wide range of functionalities out of the box , here an overview of some of the most commonlu used moduled and packages in the python Standard library.

```python
import array

arr= array.array('i',[1,2,3,4])
print(arr)
```

```python
import math 

print(math.sqrt(16))
print(math.pi)
```

```python
import random
print(random.randint(1,10))  # it will select any number between 1 to 10

print(random.choice(['apple','banana','cherry','dragon','gauva']))
```

```python
## Data Serialization

import json
data= {'name':'kajal','age':28,'height':5.4}

json_str=json.dumps(data)
print(json_str)
print(type(json_str))




parsed_Data =json.loads(json_str)
print(parsed_Data)
print(type(parsed_Data))
 
```

```python
# csv

import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Kajal',28])

with open('example.csv',mode='r') as file:

    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
# date time

from datetime import datetime, timedelta


now=datetime.now()
print(now)


yesterday= now-timedelta(days=1)
print(yesterday)
```

```python
# time

import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
#  regular expression


import re

pattern=r'\d+'
text='there are 123 apples 456'
match= re.search(pattern,text)
print(match.group())
```
