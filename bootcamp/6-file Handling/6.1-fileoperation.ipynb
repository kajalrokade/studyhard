File Operation : Read and WRite

```python
## REad a whole file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# Read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # strip() removes the new line character
        
```

```python
## Writing a file (Overwriting)

with open('example.txt','w') as file:
    file.write("Hello World !\n")
    file.write('This is the new line')
```

```python
## Write a file (without overwriting)

with open('example.txt','a') as file:
    file.write("Append operation taking place\n")
    file.write('This is the new line')
```

```python
## Wrting a list of lines to a file


lines=["Fisrt line\n","second line\n","Third line \n","Fourth line\n"]
with open('example.txt','a') as file:
    file.writelines(lines)

```

```python
## Write in Binary Files 

data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## Read from Binary Files 

with open('example.bin','rb') as file:
    print(file.read())
    
```

```python
## REad the content from a source text file and write tot the destingation text file


with open('source.txt','r') as source_file:
    content=source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
## Read text file and count the number of lines words and characters

def count_text(file_path):
    with open(file_path,'r') as file:
        lines= file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_Count=sum(len(line) for line in lines)
    return line_count,word_count,char_Count



file_path='example.txt'
lines,words,characters=count_text(file_path)

print(f'Lines : {lines}, Words:{words} , Characters: {characters}')
```

```python
## Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello WOrld\n")
    file.write("This is a new line \n")

    # move the file cursor to the begining
    file.seek(0)
    # Read the content of file
    content=file.read()
    print(content)

```
